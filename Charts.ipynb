```r
 # install.packages("vip")
library(broom)
library(tidyverse)
library(dplyr)
library(tidymodels)
library(ranger)
library(readr)       
library(janitor)
library(scales)
library(ggplot2)
library(vip)
```

```r
dataset_clean <- dataset %>%
  filter(Price > 0 & Price < 500000) 

dataset_clean <- dataset_clean %>%
                    filter(Runned_Miles < 20000)

dataset_clean <- dataset_clean %>%
                    filter(Door_num <= 5)

dataset_clean <- dataset_clean %>% 
                    filter(repair_hours > 0 & repair_hours <= 50)
```

```r
my_plot <- ggplot(dataset, aes(x = Maker)) +
  geom_bar() +
  labs(title = "Distribution of Vehicle Brands",
       x = "Brand",
       y = "Number of Vehicles") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1, size = 10))

print(my_plot)
```

```r
fuel_counts <- dataset_clean %>%
  count(Fuel_type)  

ggplot(fuel_counts, aes(x = "", y = n, fill = Fuel_type)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar(theta = "y") +  
  labs(title = "Distribution of Fuel Types",
       x = NULL, y = NULL) +
  theme_void()
```

```r
my_plot <- ggplot(dataset_clean, aes(x = Price)) +
  geom_histogram(binwidth = 1000, fill = "skyblue", color = "black") +
  labs(title = "Distribution of Vehicle Prices",
       x = "Price",
       y = "Number of Vehicles") +
  theme_minimal() +
  scale_x_continuous(labels = scales::comma, limits = c(0, 100000))  

print(my_plot)
```

```r
my_plot <- ggplot(dataset_clean, aes(x = factor(Door_num))) + 
  geom_bar(fill = "lightcoral", color = "black") + 
  labs(title = "Distribution of the Number of Vehicle Doors",
       x = "Number of Doors",
       y = "Number of Vehicles") +
  theme_minimal()

print(my_plot)
```

```r
my_plot <- ggplot(dataset_clean, aes(x = Bodytype, y = Price)) +
  geom_boxplot(fill = "lightblue", color = "black", outlier.alpha = 0.2) +
  labs(title = "Distribution of Prices by Body Type",
       x = "Body Type",
       y = "Price") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  scale_y_continuous(labels = scales::comma)

print(my_plot)
```

```r
my_plot <- ggplot(dataset_clean, aes(x = issue, y = repair_cost, color = Gearbox)) +
  geom_jitter(width = 0.3, alpha = 0.6, size = 1.5) +
  labs(title = "Repair Cost by Issue Type and Gearbox Type",
       x = "Issue Type",
       y = "Repair Cost (RON)",
       color = "Gearbox Type") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  scale_y_continuous(labels = scales::comma)

print(my_plot)
```

```r

ggplot(dataset_clean %>% filter(Gearbox %in% c("Automatic", "Manual")), aes(x = repair_hours, fill = Gearbox)) +
  geom_histogram(position = "identity", alpha = 0.3, bins = 50) +
  labs(title = "Repair Hours by Gearbox",
       x = "Repair Hours", y = "Count") +
  theme_minimal()
```

```r
ggplot(dataset_clean %>% filter(Gearbox %in% c("Automatic", "Manual")), 
       aes(x = repair_hours)) +
  geom_histogram(bins = 30, fill = "darkorange", color = "white") +
  facet_wrap(~ Gearbox) +
  labs(title = "Repair Hours by Gearbox",
       x = "Repair Hours", y = "Count") +
  theme_minimal()

```
